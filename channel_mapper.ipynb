```python
import keysight.qcs as qcs

n_qubits = 2

# define virtual channels
awgs = qcs.Channels(range(n_qubits), "xy_channels", True)
readout_awgs = qcs.Channels(range(n_qubits), "readout_channels", True)
dig = qcs.Channels(range(n_qubits), "acquire_channels")
#flux = qcs.Channels(range(n_qubits), "flux_channels")

# define the address of physical channels
awg_1x4x2 = qcs.Address(1, 4, 2)
dig_1x3x2 = qcs.Address(1, 3, 2)
dc_1x2x2 = qcs.Address(1, 2, 2)
#flux_axbxc = qcs.(Address(a, b, c))

# initialize a mapper and specify the mapping
mapper = qcs.ChannelMapper()
mapper.add_channel_mapping(awgs, [awg_1x4x2] * n_qubits, qcs.InstrumentEnum.M5300AWG)
mapper.add_channel_mapping(readout_awgs, [awg_1x4x2] * n_qubits, qcs.InstrumentEnum.M5300AWG)
mapper.add_channel_mapping(dig, [dig_1x3x2] * n_qubits, qcs.InstrumentEnum.M5200Digitizer)
#mapper.add_channel_mapping(flux, [flux_axbxc] * n_qubits, qcs.InstrumentEnum.M5301AWG)

# specify the downconverter for the digitizer at `dig_1x3x4`
mapper.add_downconverters(dig_1x3x2, dc_1x2x2)

mapper.set_delays(dig_1x3x2, 160e-9)

# set the LO frequency of the AWG and downconverter channels to be the same value by
# passing in a range of possible RF frequencies
mapper.add_downconverters(dig_1x3x2, dc_1x2x2)
mapper.constrain_lo_frequencies([awg_1x4x2, dc_1x2x2], 5.5e9, 6.4e9)

qcs.save(mapper, "channel_mapper.qcs")
```
